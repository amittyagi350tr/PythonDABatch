### groupby operations

```python
import pandas as pd
import plotly.express as pe
```

```python
link = "https://raw.githubusercontent.com/techficent/PythonDABatch/main/titanic.csv"
df = pd.read_csv(link)
df
```

### Average is called mean
### Total is called sum

aggregation : Mean, sum, min & max
don't use all data together

divide the data into sections and calculate aggregation separately

sections must be categorical

### step 1: select column/columns for groupby
### step 2: select the column with values for aggregation
### step 3: choose aggregation

### write the query now!

```python
cols = [    "Pclass"    ]

values = [ "Age"  ]

operations = [ 'min', 'max' ]


df.groupby(      cols      )   [      values      ].agg(     operations      )
```

```python
df.groupby(      [    "Pclass"    ]      )   [      [ "Age"  ]      ].agg(     [ 'min', 'max' ]      )
```

# Group by Pclass and Gender to find average Fare

```python
cols = [ "Pclass", "Sex"  ]

values = ["Fare"]

operations = ["mean"]


df.groupby(   cols   ) [  values  ].agg( operations  )
```

```python
df.groupby(   [ "Pclass", "Sex"  ]   ) [  ["Fare"]  ].agg( ["mean"]  )
```

### How many data points are present in each group when grouping is according to Pclass?

```python
df.groupby(    ['Pclass']    ).size()
```

## after grouping by Survived column, find few rows from each group (for sampling)

```python
df.groupby(    ['Survived']     ).head(2)
```

```python
df.groupby(    ['Survived']     ).sample(2)
```

### find avergae, min & max of Age & Fare, for data grouped by Pclass, Gender, Survived & Embarked columns

```python
cols = ["Pclass","Sex", "Survived", "Embarked"]

values = ["Age","Fare"]

operations = ["min","max", "mean"]

df.groupby(    cols    )[    values   ].agg(      operations     )

```

### conditions: 

for e.g:  find data for passengers with over 20 years age!
        find data for passengers who paid fare between 100-200 AND are female passengers


### steps 
            a) condition should be defined
            b) apply the condition

```python
condition1 = (    df.Age   >   20    ) #writing condition

df.loc[      condition1         ] #location rows with matching condition
```

## find Name & Age for passengers who paid fare between 100-200 AND are female passengers


```python
condition1=(   df.Sex == "female"     )

condition2 = (    df.Fare.between(100,200)           )


df.loc[      condition1  &   condition2 ,     ['Name', "Age"]             ]
```

### Names of passengers who are travelling in Pclass 1 OR people who paid over 200 Fare

```python
condition1=(df.Pclass == 1)
condition2=(df.Fare > 200)

df[condition1   |   condition2  ]
```

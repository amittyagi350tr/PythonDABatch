```python
import pandas as pd
import plotly.express as pe
```

```python
path = r"/home/harshit/Desktop/PythonDA/datasets/AAPL.csv"

df = pd.read_csv(path, index_col="Date", parse_dates=True)

df #YYYY-MM-DD
```

### resample by Month and calculate average close price

```python
#resample according to month

temp=df.resample("M")  [  ["Close"]   ].mean()

#how to find year and month_name

years = temp.index.year

months= temp.index.month_name()

temp.index=[ f"{y}_{m}"   for y,m  in zip(years,months) ]  #y_m

temp
```

### resample by Quarter and calculate average close price

```python
temp = df.resample("Q")[ ["Close"]  ].mean()

years=temp.index.year
quarters = temp.index.quarter

temp.index=[ f"{y}_Q{q}" for y,q in zip(years,quarters) ]

temp
```

```python
for time in ["Q","M"]:
   
    #common step
    temp=df.resample(time)  [  ["Close"]   ].mean()
    p1 = temp.index.year

    #specific for each
    if time == "Q":
        p2 = temp.index.quarter

    if time == "M":
        p2= temp.index.month_name()

    temp.index=[ f"{x}_{y}"   for x,y  in zip(p1,p2) ]  #y_m


    display(temp)
```

```python
with pd.ExcelWriter("apple_data_resample.xlsx") as writer:

    #run a loop for M, Q, Y & W
    for time in ["M","Q","Y","W"]:

        #resample according to 'time' period and save to a temp variable
        temp = df.resample(time) [  ["Close"]  ].mean()

        if time == "M":
            """
                make a list of year & month name where year & month_name come from temp.index zipped together
            """

            temp.index =  [    f"{year}_{month_name}" for year,month_name in zip(temp.index.year, temp.index.month_name())         ]

        if time == "Q":
            """
                make a list of year & quarter where year & quarter come from temp.index zipped together
            """
            temp.index =  [    f"{year}_{quarter}" for year,quarter in zip(temp.index.year, temp.index.quarter)         ]

        if time == "Y":
            temp.index =  [    f"{year}" for year in temp.index.year        ]

        if time == "W":
            temp.index =  [    f"{year}_{week}" for year,week in zip(temp.index.year, temp.index.week)         ]

        #save output to a page in opened excel file
        temp.to_excel(excel_writer=writer, sheet_name = time)
```

### JOIN QUERIES

```python
d1 = {
    "city" : ["Mumbai","Delhi","New York"],
    "temperature" :[30, 43, 18]
}


d2={
    "city" :["Delhi","London","Mumbai"],
    "humidity" : [40, 20, 55]
}


df1=pd.DataFrame(d1)

df2=pd.DataFrame(d2)
```

```python
#common entries for city have been merged
df1.merge(df2, on="city", how="inner"  ) 
```

```python
#common entries for city + any extra city in df1
# have been merged
df1.merge(df2, on="city", how="left"  ) 
```

```python
#common entries for city + any extra city in df2
# have been merged
df1.merge(df2, on="city", how="right"  ) 
```

```python
#common entries for city + any extra city in df1 & df2
# have been merged
df1.merge(df2, on="city", how="outer"  ) 
```

```python
#with result.xlsx file opened with name "writer", do the following
with pd.ExcelWriter("result.xlsx") as writer:

    #run a loop on all types of merge
    for task in ["inner","left","right","outer"]:

        #perform one type of merge
        temp = df1.merge(df2, on='city',how=task)

        #save output to a page in opened excel file
        temp.to_excel(excel_writer=writer, sheet_name = task)
```

```python
with pd.ExcelWriter("apple_data_resample.xlsx") as writer:

    #run a loop for M, Q, Y & W
    for time in ["M","Q","Y","W"]:

        #resample according to 'time' period and save to a temp variable
        temp = df.resample(time) [  ["Close"]  ].mean()

        if time == "M":
            """
                make a list of year & month name where year & month_name come from temp.index zipped together
            """

            temp.index =  [    f"{year}_{month_name}" for year,month_name in zip(temp.index.year, temp.index.month_name())         ]

        if time == "Q":
             """
                make a list of year & quarter where year & quarter come from temp.index zipped together
            """
            temp.index =  [    f"{year}_{quarter}" for year,quarter in zip(temp.index.year, temp.index.quarter)         ]

        if time == "Y":
            temp.index =  [    f"{year}" for year in temp.index.year        ]

        if time == "W":
            temp.index =  [    f"{year}_{week}" for year,week in zip(temp.index.year, temp.index.week)         ]

        #save output to a page in opened excel file
        temp.to_excel(excel_writer=writer, sheet_name = time)
```

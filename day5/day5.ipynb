```python
import pandas as pd
```

```python
data = {
    "Id" :  [1,2,3,4,5],

    "Name"  : ["Harshit", "Pranali", "Rupa", "Rohan", "Ajay"],

    "Age" : [32, 24, 30, 29, 45],

}

temp=pd.DataFrame(     data    )

temp
```

```python
temp['Salary'] = [   25000, 50000, 60000, 50000, 30000    ]

temp
```

```python
temp['Tax'] = (10/100) * temp['Salary']
temp
```

```python
temp['Package'] = temp['Salary'] + temp['Tax']
temp
```

```python
emp = { 
    "Id" : 6,
    "Name" : "Vijay",
    "Age" : 28,
    "Salary" : 40000,
    "Tax" : 4000,
    "Package" : 44000
}


temp =   temp.append(emp, ignore_index=True)

temp
```

## Calculate the average LoanAmount for each category separately FOR each categorical column

```python
# use a loop over all categorical columns
for col in categorical_columns:

    #average loan amount for categories in the current column
    data = df.groupby(   [col]   )[     ['LoanAmount']    ].mean()

    #rename of the column
    data = data.rename(    columns={"LoanAmount" : "AverageLoanAmount"}   )

    #create the figure: bar chart
    fig = pe.bar(data_frame=data,  x=data.index,   y='AverageLoanAmount', color=data.index)

    #fix the x axes so it looks like separate categories
    fig.update_xaxes(type='category')

    #display fig
    display(fig)
```
